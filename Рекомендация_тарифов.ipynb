### DecisionTreeClassifier - –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

````python
from sklearn.metrics import precision_score, recall_score

#–¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π
best_depth = 0
best_accuracy_dtc = 0
best_model = None
accuracy_list, best_depth_list, accuracy_train_list = [], [], []
for depth in range(1, 31):
    model = DecisionTreeClassifier(random_state = 12345, max_depth = depth) # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å –∑–∞–¥–∞–Ω–Ω–æ–π –≥–ª—É–±–∏–Ω–æ–π –¥–µ—Ä–µ–≤–∞
    model.fit(features_train, target_train)
    accuracy = accuracy_score(target_valid, model.predict(features_valid))
    accuracy_train = accuracy_score(target_train, model.predict(features_train))
    best_depth_list.append(depth)
    accuracy_list.append(accuracy)
    accuracy_train_list.append(accuracy_train)
    if accuracy > best_accuracy_dtc:
        best_depth = depth
        best_accuracy_dtc = accuracy
        best_model = model
precision = precision_score(target_valid, best_model.predict(features_valid))
recall = recall_score(target_valid, best_model.predict(features_valid))
print("–õ—É—á—à–∏–π –ø—Ä–æ—Ö–æ–¥ —Å –≥–ª—É–±–∏–Ω–æ–π:", best_depth, " –ø—Ä–∏ –ª—É—á—à–µ–º –∑–Ω–∞—á–µ–Ω–∏–∏ accuracy_score:", best_accuracy_dtc)
print("–¢–æ—á–Ω–æ—Å—Ç—å (—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–µ –º–∞—Ä–∫–∏—Ä–æ–≤–∞—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –æ–±—Ä–∞–∑–µ—Ü –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π):", precision, "—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–∞–π—Ç–∏ –≤—Å–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—Ä–∞–∑—Ü—ã:", recall)
````

````python
# –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞ –Ω–∞ –æ–±—É—á–∞—é—â–∏—Ö –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
plt.figure(figsize=(10, 5))
plt.plot(best_depth_list, accuracy_list, label='valid')
plt.plot(best_depth_list, accuracy_train_list, label='train')
plt.scatter(best_depth, best_accuracy_dtc)
plt.annotate(f'{best_accuracy_dtc=:.4f}', xy=(best_depth, best_accuracy_dtc), xytext=(-16, 10), textcoords='offset points', fontweight='bold')
plt.legend(loc='center right')
plt.xlabel('–≥–ª—É–±–∏–Ω–∞ max_depth')
plt.ylabel('—Ç–æ—á–Ω–æ—Å—Ç—å accuracy')
plt.title('–ú–æ–¥–µ–ª—å DecisionTreeClassifier')
plt.show()
````

–†–∞–±–æ—Ç—É –∞–ª–≥–æ—Ä–∏—Ç–º–∞ ```DecisionTreeClassifier``` –º–æ–∂–Ω–æ –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—Ç—å –ø–æ –º–Ω–æ–∂–µ—Å—Ç–≤—É –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º—ã—Ö –∫—Ä–∏—Ç–µ—Ä–∏–µ–≤ –æ—Å—Ç–∞–Ω–æ–≤–∞, —Ç–∞–∫–∏—Ö –∫–∞–∫:
- –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞
- –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ª–∏—Å—Ç—å–µ–≤
- –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ n-–æ–±—ä–µ–∫—Ç–æ–≤ –≤ –ª–∏—Å—Ç–µ
- –æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –≤ –ª–∏—Å—Ç–µ –æ–∫–∞–∑–∞–ª–∏—Å—å –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞
- –∏ –¥–∞–∂–µ –æ—Ç–∫–∞—Ç –Ω–∞–∑–∞–¥ –ø—Ä–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–º –∫–∞—á–µ—Å—Ç–≤–µ —Ä–∞–∑–±–∏–µ–Ω–∏—è –Ω–∞ –∫–∞–∫—É—é-—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –≤–µ–ª–∏—á–∏–Ω—É n-–æ–±—ä–µ–∫—Ç–æ–≤

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã—Å—á–∏—Ç—ã–≤–∞–µ—Ç –≤ –ª–∏—Å—Ç—å—è—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –æ–±—ä–µ–∫—Ç–∞ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –∫–ª–∞—Å—Å–æ–≤ (–∑–¥–µ—Å—å —ç—Ç–æ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ç–µ–ø–µ–Ω—å—é —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏) –∏ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —Å–≤–æ–µ–π —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–µ–ª–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –Ω–æ–≤—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤. –ï—Å–ª–∏ ```DecisionTreeClassifier``` –Ω–∏–∫–∞–∫ –Ω–µ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞—Ç—å, —Ç–æ –¥–µ—Ä–µ–≤–æ –Ω–∞—É—á–∏—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ —Ç–æ—á–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Ç–µ—Ä—è–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –∞–¥–µ–∫–≤–∞—Ç–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å –º–µ—Ç–∫–∏ –Ω–æ–≤—ã—Ö –ø—Ä–∏—Ö–æ–¥—è—â–∏—Ö –∫ –Ω–µ–º—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Ç.–æ. —Å–ª—É—á–∏—Ç—Å—è *–ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ* (overfitting). –ò–º–µ–Ω–Ω–æ –µ–≥–æ –º—ã –º–æ–∂–µ–º –≤–∏–¥–µ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤—ã—à–µ - —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–µ—Ä–µ–≤—å–µ–≤ –∑–Ω–∞—á–µ–Ω–∏–µ accuracy –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –ø–æ—Å—Ç–æ—è–Ω–Ω–æ —Ä–∞—Å—Ç—ë—Ç, —Ç–æ–≥–¥–∞ –∫–∞–∫ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ–Ω–æ –ø–∞–¥–∞–µ—Ç. –∞ –º—ã –±–µ—Ä–µ–º –ª—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ *accuracy* –∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö - —ç—Ç–æ:

$accuracy = 0.802$

### RandomForestClassifier - —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

````python
#—Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å
best_est = 0
best_forest_depth = 0
best_accuracy_rfc = 0
best_split = 0
forest_best_model = None
best_scores = []
for est in range(10, 51, 10):
    for depth in range(1, 31):
        best_split_list, accuracy_list_rfc, accuracy_train_list_rfc = [], [], []
        for split in range(2, 32):
            forest_model = RandomForestClassifier(random_state = 12345, n_estimators = est, max_depth = depth, min_samples_split=split) # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å –∑–∞–¥–∞–Ω–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ä–µ–≤—å–µ–≤
            forest_model.fit (features_train, target_train)
            accuracy_rfc = accuracy_score(target_valid, forest_model.predict(features_valid))
            accuracy_train_rfc  =  accuracy_score(target_train, forest_model.predict(features_train))
            accuracy_list_rfc.append(accuracy_rfc)
            accuracy_train_list_rfc.append(accuracy_train_rfc)
            best_split_list.append(split)
            if accuracy_rfc > best_accuracy_rfc:
                best_forest_depth = depth
                best_est = est
                best_split = split
                best_accuracy_rfc = accuracy_rfc #  —Å–æ—Ö—Ä–∞–Ω–∏–º –Ω–∞–∏–ª—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ accuracy –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                forest_best_model = forest_model
        best_scores.append((best_accuracy_rfc, accuracy_train_list_rfc, accuracy_list_rfc))

# –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞ –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –ø—Ä–æ—Ö–æ–¥ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º best_accuracy
best_accuracy_rfc, best_train_accuracy, best_valid_accuracy = max(best_scores, key=lambda x: x[0])

rfc_precision = precision_score(target_valid, forest_best_model.predict(features_valid))
rfc_recall = recall_score(target_valid, forest_best_model.predict(features_valid))
print("Accuracy –Ω–∞–∏–ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:", best_accuracy_rfc, '–ø—Ä–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è:', best_split, ' –∏ –ø—Ä–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –¥–µ—Ä–µ–≤—å–µ–≤:', best_est, '—Å –≥–ª—É–±–∏–Ω–æ–π:', best_forest_depth)
print("–¢–æ—á–Ω–æ—Å—Ç—å (—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–µ –º–∞—Ä–∫–∏—Ä–æ–≤–∞—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –æ–±—Ä–∞–∑–µ—Ü –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π):", rfc_precision, "—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–∞–π—Ç–∏ –≤—Å–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—Ä–∞–∑—Ü—ã:", rfc_recall)
````

````python
# –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞ –Ω–∞ –æ–±—É—á–∞—é—â–∏—Ö –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
plt.figure(figsize=(10, 5))
plt.plot(best_split_list, best_valid_accuracy, label='valid')
plt.plot(best_split_list, best_train_accuracy, label='train')
plt.scatter(best_split, best_accuracy_rfc)
plt.annotate(f'{best_accuracy_rfc=:.4f}', xy=(best_forest_depth, best_accuracy_rfc), xytext=(-16, 10), textcoords='offset points', fontweight='bold')
plt.legend(loc='center right')
#plt.xticks(range(len(best_valid_accuracy)), range(1, len(best_valid_accuracy)), rotation=90)
plt.xlabel('–º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è')
plt.ylabel('—Ç–æ—á–Ω–æ—Å—Ç—å accuracy')
plt.title('–ú–æ–¥–µ–ª—å RandomForestClassifier')
plt.show()
````

```RandomForestClassifier``` - —ç—Ç–æ –æ–¥–Ω–∞ –∏–∑ —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç–µ–π  –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤, –∏—Å–ø–æ–ª—å–∑—É—é—â–∏—Ö –≤ —Å–≤–æ–µ–π –æ—Å–Ω–æ–≤–µ —Ç—É –∂–µ –ª–æ–≥–∏–∫—É —Ä–∞–∑–±–∏–µ–Ω–∏—è –Ω–∞ —É–∑–ª—ã –∏ –ª–∏—Å—Ç—å—è, —á—Ç–æ –∏ ```DecisionTree```-–∞–ª–≥–æ—Ä–∏—Ç–º—ã, —Ç–æ–ª—å–∫–æ –∫–æ–Ω–µ—á–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –∑–¥–µ—Å—å –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è –Ω–µ –ø—Ä–æ—Å—Ç–æ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —Å—Ç–µ–ø–µ–Ω–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≥–æ –∫–∞–∫–æ–≥–æ-–ª–∏–±–æ –¥–µ—Ä–µ–≤–∞, –∞ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è —Ç.–Ω. *–≥–æ–ª–æ—Å–æ–≤–∞–Ω–∏–µ –¥–µ—Ä–µ–≤—å–µ–≤*, —Ç–∞–∫ –∫–∞–∫ –≤ –∞–ª–≥–æ—Ä–∏—Ç–º–µ –∏—Ö –º–æ–∂–µ—Ç –±—ã—Ç—å –º–Ω–æ–≥–æ.

–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–µ—Ä–µ–≤—å–µ–≤ –≤ –ª–µ—Å—É —Ç–∞–∫–∏–µ –∂–µ —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–Ω—ã–µ, –∫–∞–∫ –∏ –¥–ª—è –æ–¥–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞, –∏ –≤—Å–µ –æ–Ω–∏ –º–æ–≥—É—Ç –≤–ª–∏—è—Ç—å –Ω–∞ –∫–æ–Ω–µ—á–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è.

–ì–ª–∞–≤–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ ```RandomForestClassifier``` –≤ —Ç–æ–º, —á—Ç–æ –∫–∞–∂–¥–æ–µ –¥–µ—Ä–µ–≤–æ —Å—Ç—Ä–æ–∏—Ç—Å—è –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ª—É—á–∞–π–Ω–æ–π –ø–æ–¥–≤—ã–±–æ—Ä–∫–∏ –æ–±—ä–µ–∫—Ç–æ–≤ –∏ —Å–ª—É—á–∞–π–Ω–æ–π –ø–æ–¥–≤—ã–±–æ—Ä–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —á—Ç–æ —Å–≤–æ–¥–∏—Ç –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–∞ –Ω–µ—Ç *–ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ*, —Å–Ω–∏–∂–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É —Ä–∞–∑–±—Ä–æ—Å–∞ –¥–∞–Ω–Ω—ã—Ö (–∫–æ—Ç–æ—Ä–æ–π –±–æ–ª–µ—é—Ç –¥–µ—Ä–µ–≤—å—è - —ç—Ç–æ –∫–æ–≥–¥–∞ –ø—Ä–∏ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–º –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –ø–æ–ª—É—á–∞–µ—Ç—Å—è —Å–æ–≤—Å–µ–º –¥—Ä—É–≥–æ–µ –¥–µ—Ä–µ–≤–æ) –∏ –≤ —Ü–µ–ª–æ–º —É–ª—É—á—à–∞–µ—Ç –æ–±–æ–±—â–∞—é—â—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏. –î–ª—è –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö *accuracy*:

$accuracy = 0.824$

### LogisticRegression - –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

````python
#–ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è
accuracy_list_lr, accuracy_train_list_lr, iter_list = [], [], []
best_accuracy_lr, accuracy_train_lr, n_iter = 0, 0, 0
best_logistic_model = None
for i in range(1, 101):
    model_logistic = LogisticRegression(random_state=12345, max_iter=i)
    model_logistic.fit(features_train_scaled, target_train)
    accuracy_lr = accuracy_score(target_valid, model_logistic.predict(features_valid_scaled))
    accuracy_train_lr  =  accuracy_score(target_train, model_logistic.predict(features_train_scaled))
    accuracy_list_lr.append(accuracy_lr)
    accuracy_train_list_lr.append(accuracy_train_lr)
    iter_list.append(i)
    if accuracy_lr > best_accuracy_lr:
        best_accuracy_lr = accuracy_lr
        best_logistic_model = model_logistic
        n_iter = i
lr_precision = precision_score(target_valid, best_logistic_model.predict(features_valid_scaled))
lr_recall = recall_score(target_valid, best_logistic_model.predict(features_valid_scaled))

print("Accuracy –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:", best_accuracy_lr, ',–ø—Ä–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∏—Ç–µ—Ä–∞—Ü–∏–π =', n_iter)
print("–¢–æ—á–Ω–æ—Å—Ç—å (—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–µ –º–∞—Ä–∫–∏—Ä–æ–≤–∞—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –æ–±—Ä–∞–∑–µ—Ü –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π):", lr_precision, "—Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–∞–π—Ç–∏ –≤—Å–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—Ä–∞–∑—Ü—ã:", lr_recall)
````
